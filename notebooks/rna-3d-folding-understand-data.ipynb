# Introduction

---

## What is RNA?
RNA (Ribonucleic Acid) is a **nucleic acid**, a type of biological macromolecule, similar to DNA but with key differences in structure and function. It plays a crucial role in genetics and cellular biology.

### Differences between RNA and DNA
| Feature          | DNA                      | RNA                      |
|-----------------|--------------------------|--------------------------|
| Sugar           | Deoxyribose               | Ribose                   |
| Strands         | Double-stranded (mostly, forming a **double helix**)  | Single-stranded (mostly) |
| Bases          | A, T, C, G                 | A, **U**, C, G (Uracil replaces Thymine) |
| Stability       | More stable               | Less stable (prone to degradation) |
| Function        | **Genetic storage, replication** (stores hereditary information in cells) | **Protein synthesis, regulation, catalysis** (actively involved in gene expression) |

### **DNA as the Blueprint of Life**
DNA is **the genetic material in all known living organisms**. It carries the instructions for the growth, development, functioning, and reproduction of cells. The **double-helix structure** ensures that genetic information is stored safely and can be accurately copied during **cell division** (mitosis and meiosis). Every cell in an organism contains the same DNA, which acts as the "blueprint" for making proteins that define cellular function.

---

## **Types of RNA**
RNA comes in different forms, each with distinct functions that work together to **convert genetic information into proteins**. The three main types of RNA directly involved in **protein synthesis** are:

### **1. Messenger RNA (mRNA) – The Genetic Courier**
- **Function:** Carries genetic instructions from DNA to ribosomes to synthesize proteins. It is a temporary copy of a gene used to direct protein synthesis.
- **Example:** **mRNA vaccines** (e.g., Pfizer-BioNTech COVID-19 vaccine) work by introducing a synthetic mRNA sequence that cells use to produce a viral protein, training the immune system to recognize and fight the virus.

### **2. Transfer RNA (tRNA) – The Adapter Molecule**
- **Function:** Brings amino acids to the ribosome, matching them with the corresponding mRNA codon.
- **Structure:** It has a **cloverleaf** shape and an **anticodon region** that pairs with mRNA codons.
- **Example:** If mRNA has the codon `AUG` (Methionine), tRNA carries the corresponding amino acid (Methionine) and adds it to the growing protein chain.

### **3. Ribosomal RNA (rRNA) – The Protein Factory**
- **Function:** Forms the core of ribosomes and catalyzes peptide bond formation between amino acids.
- **Highly conserved:** Used in evolutionary studies to compare species.
- **Example:** The ribosome consists of both rRNA and proteins, acting as the **assembly site for proteins**.

### **4. Small Nuclear RNA (snRNA) – The Gene Editor**
- **Function:** Involved in **RNA processing**, especially **splicing**, which removes non-coding sequences (**introns**) from pre-mRNA before it becomes mature mRNA.
- **Example:** Helps form the **spliceosome**, a molecular machine that cuts out introns and joins exons together, allowing different proteins to be produced from the same gene.

### **5. MicroRNA (miRNA) & Small Interfering RNA (siRNA) – The Gene Silencers**
- **Function:** Regulate gene expression by binding to mRNA and preventing translation or promoting degradation.
- **Example:** miRNAs are involved in **cancer suppression** by regulating oncogenes (cancer-causing genes). siRNA is used in **gene therapy** to silence disease-causing genes.

---

## **Why is RNA important?**
RNA is fundamental in biology and medicine because it acts as a bridge between genetic information (DNA) and functional molecules (proteins).

### **1. Role in Protein Synthesis**
RNA is central to the **central dogma of molecular biology**:

**DNA → RNA → Protein**

- **Transcription:** DNA is transcribed into mRNA in the nucleus.
- **Translation:** mRNA is translated into proteins by ribosomes (rRNA + tRNA) in the cytoplasm.
- **Example of the process:**
  1. DNA: `TAC GGC TAA`
  2. mRNA: `AUG CCG AUU`
  3. tRNA brings the corresponding amino acids: Methionine - Proline - Isoleucine
  4. The ribosome links these amino acids together to form a protein.

### **2. Gene Regulation & Cell Function**
RNA is not just a passive messenger; it actively regulates gene expression.
- **miRNA and siRNA** silence genes by binding to mRNA and blocking translation.
- **Long non-coding RNA (lncRNA)** influences chromatin structure and gene expression.
- **Riboswitches** change their conformation to regulate gene activity based on environmental conditions.

### **3. Importance in Evolution & Medicine**
- RNA was likely the **first genetic material** in early life forms (**RNA world hypothesis**).
- Many viruses, including **SARS-CoV-2**, use RNA as their genetic material.
- **RNA-based drugs** and **RNA vaccines** are revolutionizing medicine.

---
## **What is 3D RNA folding?**
Unlike DNA, RNA does not stay linear. It **folds into complex structures** that determine its function.

### **Levels of RNA Structure**
1. **Primary Structure**: The sequence of nucleotides (A, U, C, G).
2. **Secondary Structure**: Local folding patterns (stems, loops, bulges).
3. **Tertiary Structure**: The full **3D shape** stabilized by interactions.

### **Key Folding Elements**
- **Hairpin loops**: Short double-stranded stems with a loop at the end.
- **Pseudoknots**: Complex folds where loops interact with other RNA regions.
- **Coaxial stacking**: Stacking of helices for stability.

RNA folding is **not random**—it follows **thermodynamic principles** to find the most stable conformation.

---

## **Why do we need to study 3D RNA folding?**
The **3D structure determines RNA function**, affecting its ability to interact with other molecules.

### **1. RNA-Protein Interactions**
- Ribosomes, CRISPR, and viral replication depend on RNA-protein binding.

### **2. Drug Design & Synthetic Biology**
- Many drugs target RNA structures (e.g., antibiotics targeting rRNA).
- Artificially designed RNA molecules can be used in therapy.

---

## **How does RNA 3D folding work?**
RNA folds due to various **molecular forces** and stabilizing factors.

### **1. Key Forces in RNA Folding**
- **Hydrogen bonding**: Base pairing (A-U, G-C).
- **Stacking interactions**: π-π interactions between bases.
- **Metal ion stabilization**: Mg²⁺ ions stabilize tertiary structures.

### **2. Experimental Methods to Study RNA Folding**
- **X-ray crystallography**: High-resolution but difficult for RNA.
- **NMR spectroscopy**: Good for small RNAs.
- **Cryo-EM**: Ideal for large RNA complexes like ribosomes.

### **3. Computational Approaches**
- **RNAfold**: Predicts 2D structures based on thermodynamics.
- **Rosetta & AlphaFold RNA**: AI-driven 3D structure prediction.
- **Molecular dynamics simulations**: Simulates RNA folding over time.

---

## **Challenges & Future Directions**
Despite progress, RNA structure prediction remains a challenge.

### **1. Predicting RNA Structures is Difficult**
- Experimental methods are slow and expensive.
- Computational models still struggle with large, complex RNAs.

### **2. Combining Experimental & Computational Approaches**
- Integrating lab data (e.g., Cryo-EM) with AI models improves accuracy.

### **3. AI & Deep Learning in RNA Folding**
- **AlphaFold RNA** is emerging as a powerful tool.
- Machine learning could revolutionize RNA-based drug discovery.

---
### **Modern Approaches in RNA Structure Prediction**
Predicting RNA 3D structures is a complex challenge that combines experimental and computational techniques. In recent years, **AI-driven models** have significantly improved RNA structure prediction.

- **RNAfold**: A widely used thermodynamic model that predicts RNA secondary structures based on minimum free energy.
- **Rosetta**: A computational tool that models RNA tertiary structures using fragment-based assembly and molecular dynamics.
- **AlphaFold RNA**: A deep learning-based approach, inspired by AlphaFold for proteins, aiming to predict RNA 3D structures with high accuracy.

By integrating **machine learning, molecular simulations, and experimental data**, these tools are paving the way for more precise RNA structural insights, benefiting **drug discovery, synthetic biology, and gene regulation studies**.

### **Example: RNAfold in Drug Discovery**
One practical application of RNA structure prediction is in **drug discovery**. Many **antiviral drugs** target RNA structures to disrupt viral replication. For example, **RNAfold** has been used to predict the secondary structures of **SARS-CoV-2 RNA elements**, helping researchers identify potential drug-binding sites. This approach aids in the design of **RNA-targeted therapeutics**, which could be crucial for treating RNA viruses and genetic disorders.



## **Final Thoughts**
Understanding RNA and its 3D folding is crucial for biology, medicine, and technology. By combining experimental and computational approaches, researchers can unlock new possibilities in **drug design, genetic engineering, and synthetic biology**.

```python
def find_atoms_in_region(df, x_target, y_target, z_target, tolerance=100):
    """
    Filter all RNA atoms that are within a given tolerance range around a target coordinate.
    
    Parameters:
    df (DataFrame): DataFrame containing atomic coordinates (must have 'x', 'y', 'z' columns).
    x_target (float): X-coordinate of the region of interest.
    y_target (float): Y-coordinate of the region of interest.
    z_target (float): Z-coordinate of the region of interest.
    tolerance (float): Range around the target coordinates to search for atoms.
    
    Returns:
    DataFrame: Filtered DataFrame with atoms in the specified region.
    """
    filtered_atoms = df[
        (df["x_1"].between(x_target - tolerance, x_target + tolerance)) &
        (df["y_1"].between(y_target - tolerance, y_target + tolerance)) &
        (df["z_1"].between(z_target - tolerance, z_target + tolerance))
    ]
    return filtered_atoms

region_atoms = find_atoms_in_region(df_train_labels, -730, -100, 400)
print(region_atoms) # there you can see how i found the 4V4G. Am I right?
```

## K-mer Frequency Analysis in RNA Structure Prediction
Why look at K-mers?

RNA folding is driven by sequence patterns. K-mers—short sequence fragments—help us spot motifs that influence structure.
How does this help our competition?
- Better features for ML models
- Identifies key motifs affecting folding
- Makes predictions more interpretable
By analyzing K-mer frequencies, we capture useful patterns that could improve RNA 3D structure predictions. 

```python
# K-mer- Function, k=4
def get_kmers(sequence, k=3):
    return [sequence[i:i+k] for i in range(len(sequence) - k + 1)]

kmer_counts = Counter()
for seq in df_train_sequences["sequence"]:
    kmer_counts.update(get_kmers(seq, k=4))
print(kmer_counts.most_common(10))
```

```python
# k=8
kmer_counts = Counter()
for seq in df_train_sequences["sequence"]:
    kmer_counts.update(get_kmers(seq, k=8))

print(kmer_counts.most_common(10))
```

```python
# k=12
kmer_counts = Counter()
for seq in df_train_sequences["sequence"]:
    kmer_counts.update(get_kmers(seq, k=12))  

print(kmer_counts.most_common(10))
```

**How Do We Use K-mer Results in 3D RNA Folding?**

K-mer analysis helps us identify patterns in RNA sequences that influence 3D structure formation. But what do we do with the results?

First, we can use K-mer frequencies as features for machine learning models. By converting sequences into numerical representations, we can check if specific K-mer patterns correlate with stable or unstable 3D structures. This can improve the predictive power of our models.

Next, K-mers can reveal structural motifs. Many RNA folding elements, such as hairpins or pseudoknots, are linked to specific sequence patterns. By analyzing frequent K-mers, we might detect biologically relevant motifs that contribute to RNA stability. Comparing them with predictions from RNAfold or 3D modeling tools can help validate their importance.

Another approach is clustering sequences based on K-mer distributions. If sequences with similar K-mer profiles tend to form similar 3D structures, this could provide valuable insights into RNA folding principles.

Finally, if we have ground truth 3D structures, we can compare our K-mer results to see if specific patterns consistently appear in certain structural elements. This validation step ensures that our findings are meaningful rather than random.
Key Takeaways:

- K-mers as ML features: Improve structure prediction models.
- Identify structural motifs: Detect sequence patterns linked to RNA folding.
- Cluster similar sequences: Explore relationships between K-mers and 3D structures.
- Validate with real data: Check if observed patterns align with known structures.

By leveraging K-mer analysis, we can uncover sequence-structure relationships that enhance RNA 3D folding predictions

## GC Content Analysis  

GC content represents the proportion of **guanine (G) and cytosine (C) nucleotides** in an RNA sequence. It is an important factor in RNA stability because **G≡C base pairs have three hydrogen bonds**, making them stronger than **A=U base pairs (two hydrogen bonds)**.

**Why is GC content important?**  
- **RNA stability:** Higher GC content generally leads to more stable secondary structures.  
- **Thermodynamics:** GC-rich sequences often have lower **free energy** (more stable folds).  
- **RNA function:** Certain RNA types have characteristic GC content distributions.  

Below, we analyze the **distribution of GC content** in our dataset.

```python
# Calculate GC content
df_train_sequences["GC_content"] = df_train_sequences["sequence"].apply(
    lambda seq: (seq.count("G") + seq.count("C")) / len(seq)
)

plt.figure(figsize=(10, 5))
sns.histplot(df_train_sequences["GC_content"], bins=30, kde=True)
plt.title("GC Content Distribution")
plt.xlabel("GC Content")
plt.ylabel("Frequency")
plt.show()

print(df_train_sequences["GC_content"].describe())
```

-Most sequences are GC-rich, which suggests increased RNA stability.
- A few sequences are entirely A/U or entirely G/C, which might have special properties.
- GC content varies but mostly stays within a normal biological range (40-70%).

```python
# Merge sequence-based GC content with structural labels
df_combined = df_train_sequences.copy()
df_combined["ID"] = df_train_labels["ID"]  # Assuming IDs match
df_combined["resid"] = df_train_labels["resid"]
df_combined["x_1"] = df_train_labels["x_1"]
df_combined["y_1"] = df_train_labels["y_1"]
df_combined["z_1"] = df_train_labels["z_1"]

# Compute distance of each nucleotide from the origin (as a rough spatial feature)
df_combined["distance_from_origin"] = np.sqrt(
    df_combined["x_1"]**2 + df_combined["y_1"]**2 + df_combined["z_1"]**2
)

# Scatter plot of GC content vs. 3D structure distance
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_combined["GC_content"], y=df_combined["distance_from_origin"])
plt.xlabel("GC Content")
plt.ylabel("Distance from Origin (3D Space)")
plt.title("GC Content vs. 3D RNA Spatial Structure")
plt.show()

# Compute correlation
correlation = df_combined[["GC_content", "distance_from_origin"]].corr()
print("Correlation between GC content and 3D structure distance:\n", correlation)
```

- The correlation coefficient is -0.09868, which is very close to 0.
- A weak negative correlation suggests that GC content has little to no direct impact on the 3D spatial positioning of nucleotides.

RNA folding is driven by more than just GC content
- While GC base pairs increase stability, tertiary interactions (e.g., pseudoknots, loops) determine final 3D shape.
- Other forces (like protein interactions, metal ion binding) influence structure.

## Using RNA Secondary Structure for Machine Learning & Graph Neural Networks (GNNs)
RNA is not just a linear sequence; it folds into complex secondary and tertiary structures that determine its function. The secondary structure, which can be predicted using RNAfold, serves as an intermediate step before the full 3D folding and provides critical information about base pairings, loops, and structural stability.

This section explores how RNA secondary structure can be utilized for Machine Learning and Graph Neural Networks (GNNs):
- RNA as a Graph: Nucleotides as nodes, base pairings as edges.
- Folding energy as a feature: Lower values indicate more stable structures!
- RNA classification: Using structural features in ML models (XGBoost, Random Forest).
- GNN Training: Predicting RNA structures using Graph Convolutional Networks (GCNs).

These approaches enable a deeper analysis of RNA sequences and help identify structural patterns, which is essential for applications such as RNA-based drug development and functional RNA classification.

```python
def predict_rna_structure(sequence):
    """Predicts RNA secondary structure and free energy using RNAfold."""
    process = subprocess.run(
        ["RNAfold"], input=sequence, capture_output=True, text=True
    )
    output = process.stdout.strip().split("\n")
    if len(output) < 2:
        return None, None

    structure = output[1].split(" ")[0]  # Extract secondary structure
    energy_match = re.search(r"-?\d+\.\d+", output[1])  # Extract energy
    energy = float(energy_match.group()) if energy_match else None

    return structure, energy

def rna_to_graph(sequence, structure):
    """Converts RNA sequence and secondary structure into a graph representation."""
    G = nx.Graph()
    for i, nucleotide in enumerate(sequence):
        G.add_node(i, nucleotide=nucleotide)

    # Sequential edges
    for i in range(len(sequence) - 1):
        G.add_edge(i, i + 1)

    # Base-pairing edges
    stack = []
    for i, char in enumerate(structure):
        if char == "(":
            stack.append(i)
        elif char == ")":
            if stack:
                j = stack.pop()
                G.add_edge(i, j)

    return G

# Get RNA sequence (1SCL example)
sequence = df_train_sequences["sequence"].iloc[0]

# Predict secondary structure & energy
structure, energy = predict_rna_structure(sequence)

if structure:
    print(f"RNA Structure: {structure} | Energy: {energy} kcal/mol")

    # Convert to graph and PyTorch Geometric format
    rna_graph = rna_to_graph(sequence, structure)
    edge_index = torch.tensor(list(rna_graph.edges), dtype=torch.long).t().contiguous()
    x = torch.tensor([[ord(n)] for n in sequence], dtype=torch.float)  

    data = Data(x=x, edge_index=edge_index, energy=torch.tensor([energy], dtype=torch.float))
    print(data)
else:
    print("Error: Could not generate RNA secondary structure.")
```

The secondary structure is represented in dot-bracket notation, where:
- ( and ) indicate base-paired nucleotides.
- . represents unpaired nucleotides (loops or bulges

The energy value -9.7 kcal/mol represents the stability of the RNA fold:
- More negative values indicate a more stable RNA structure.
- Near-zero or positive values suggest an unstable or unfolded RNA.

Nodes and Edges
- x=[29, 1]: There are 29 nodes, each represented by a single feature (e.g., ASCII encoding of A, U, G, C).
- edge_index=[2, 37]: The graph contains 37 edges, indicating the number of nucleotide connections (both sequential and base-pairing).
- energy=[1]: The energy feature is stored as a graph-level attribute, which can be used for stability prediction in ML models.

```python
print(f"Total Nodes: {data.num_nodes}")
print(f"Total Edges: {data.num_edges}")
print("Edge List:")
print(data.edge_index.t().tolist())  # Print all edges to inspect them
```

- Total Nodes (29): Each nucleotide in the RNA sequence is a node in the graph.
- Total Edges (37): These include both sequential connections (linking adjacent nucleotides) and base-pairing connections (from the dot-bracket notation).

- Each pair of numbers [i, j] represents an edge between nucleotide i and nucleotide j.
- The graph contains:
    - Sequential edges (e.g., [0,1], [1,2], [2,3]), which connect adjacent nucleotides.
    - Base-pairing edges (e.g., [0,28], [1,27], [2,26]), derived from the dot-bracket structure.

This edge list forms the connectivity structure required for GNN training, allowing the model to learn from both local (sequential) and long-range (base-pairing) interactions.

**Summary: Why Is This Useful?**
- Captures RNA folding structure in a format suitable for Machine Learning.
- Encodes long-range interactions, which are critical for RNA function.
- Prepares RNA data for GNN-based classification and prediction models.

# Classic EDA Overview

```python
def data_overview(data, target):
    # Overview
    display(Markdown("## Data Overview"))
    
    display(Markdown("### General Information"))
    display(Markdown(f"- Number of rows and columns: {data.shape[0]} x {data.shape[1]}"))
    display(Markdown("- Column names:"))
    display(list(data.columns))

    display(Markdown("### Data Types & Missing Values"))
    missing = data.isnull().sum()
    dtypes = pd.DataFrame(data.dtypes, columns=["Data Type"])
    missing_df = pd.DataFrame(missing, columns=["Missing Values"])
    overview_df = dtypes.join(missing_df)
    display(overview_df.style.background_gradient(cmap="coolwarm"))

    display(Markdown("### Classic head of Data"))
    display(data.head().style.set_properties(**{"background-color": "#f5f5f5"}))

    display(Markdown("### Statistical Summary (describe)"))
    display(data.describe().T.style.background_gradient(cmap="viridis"))

    # Target variable analysis
    if target is not None:
        display(Markdown(f"## Target Variable: `{target}`"))
        sns.set_style("whitegrid")  
        sns.set_palette("viridis")   

        fig, ax = plt.subplots(1, 2, figsize=(14, 5))

        #   Absolute frequency barplot
        sns.barplot(x=data[target].value_counts().index, 
                y=data[target].value_counts(), 
                ax=ax[0])  

        ax[0].set_title("Absolute Frequency", fontsize=12, fontweight="bold")
        ax[0].set_ylabel("Count")
        ax[0].set_xlabel(target)
        ax[0].grid(axis="y", linestyle="--", alpha=0.5)  

        # Percentage distribution barplot
        sns.barplot(x=data[target].value_counts().index, 
                    y=data[target].value_counts(normalize=True), 
                    ax=ax[1])  

        ax[1].set_title("Percentage Distribution", fontsize=12, fontweight="bold")
        ax[1].set_ylabel("Percentage")
        ax[1].set_xlabel(target)
        ax[1].grid(axis="y", linestyle="--", alpha=0.5)

    

        for spine in ["top", "right"]:
            ax[0].spines[spine].set_visible(False)
            ax[1].spines[spine].set_visible(False)

    plt.tight_layout()
    plt.show()
```
